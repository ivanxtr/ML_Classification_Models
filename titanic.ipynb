# Prepare the data for traininig / data cleaning

```python
sns.heatmap(training_set.isnull(), yticklabels= False, cbar = False, cmap= 'Blues')
```

```python
# inplace to modify the memory data
training_set.drop('Cabin', axis = 1, inplace= True)
sns.heatmap(training_set.isnull(), yticklabels= False, cbar = False, cmap= 'Blues')
```

```python
training_set.drop(['Name', 'Ticket', 'Embarked'], axis= 1, inplace= True)
```

```python
sns.heatmap(training_set.isnull(), yticklabels= False, cbar = False, cmap= 'Blues')
```

```python
# Average Age
plt.figure(figsize= (15,10))
sns.boxenplot(x = 'Sex', y = 'Age', data = training_set)
```

```python
# Replace empty age values for average values
def Fill_Age(data):
  age = data[0]
  sex = data[1]
  if pd.isna(age):
    if sex == 'male':
      return 29
    else:
      return 25
  else:
    return age
```

```python
training_set['Age'] = training_set[['Age', 'Sex']].apply(Fill_Age, axis = 1)
```

```python
sns.heatmap(training_set.isnull(), yticklabels= False, cbar = False, cmap= 'Blues')
```

```python
training_set['Age'].hist(bins= 20)
```

```python
training_set.drop(['PassengerId'], axis = 1, inplace= True)
```

```python
male = pd.get_dummies(training_set['Sex'], drop_first= True)
```

```python
training_set.drop(['Sex'], axis= 1, inplace=True)
```

```python
training_set = pd.concat([training_set, male], axis= 1)
```

```python
X = training_set.drop('Survived', axis=1).values
```

```python
Y = training_set['Survived'].values
```

# Model Traininig

```python
%pip install sklearn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state = 20)
```

```python
# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

# Model Testing

```python
y_predict = classifier.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot= True, fmt='d')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict))
```

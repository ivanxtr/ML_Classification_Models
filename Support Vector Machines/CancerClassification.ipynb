```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer.keys()
```

```python
# np.c_ === concatenate data
cancer_df = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns= np.append(cancer['feature_names'], 'target'))
```

```python
sns.pairplot(cancer_df, hue= 'target', vars=['mean radius',	'mean texture',	'mean perimeter',	'mean area',	'mean smoothness'])
```

```python
sns.countplot(cancer_df['target'], label= 'count')
```

```python
sns.scatterplot(x = 'mean area', y = 'mean smoothness', hue= 'target', data= cancer_df)
```

```python
sns.scatterplot(x = 'mean texture', y = 'mean radius', hue= 'target', data= cancer_df)
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(cancer_df.corr(), annot=True)
```

```python
# training
X = cancer_df.drop(['target'], axis=1)
y = cancer_df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.20)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC()
svc_model.fit(X_train, y_train)
```

```python
# Test
y_predict = svc_model.predict(X_test)
y_predict
```

```python
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

```python
# improve the model
min_train = X_train.min()
min_train
```

```python
range_train = (X_train - min_train).max()
range_train
```

```python
X_train_scaled = (X_train - min_train) / range_train
X_train_scaled
```

```python
sns.scatterplot(x = X_train['mean area'], y = X_train['mean smoothness'], hue= y_train)
```

```python
sns.scatterplot(x = X_train_scaled['mean area'], y = X_train_scaled['mean smoothness'], hue= y_train)
```

```python
min_test = X_test.min()
range_test = (X_test - min_test).max()
X_test_scaled = (X_test - min_test) / range_test
X_test_scaled
```

```python
sns.scatterplot(x = X_test_scaled['mean area'], y = X_test_scaled['mean smoothness'], hue= y_test)
```

```python
svc_model = SVC()
svc_model.fit(X_train_scaled, y_train)
```

```python
y_predict = svc_model.predict(X_test_scaled)
y_predict
```

```python
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```
